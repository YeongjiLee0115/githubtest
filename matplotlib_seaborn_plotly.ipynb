```python
from matplotlib import pyplot as plt
import seaborn as sns
from plotly import express as px
```

```python
import numpy as np
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
#the most commonly used : matplotlib.pyplot
plt.plot(diamonds['x'], diamonds['y'])
```

```python
plt.plot(diamonds['x'])
```

```python
plt.plot(diamonds['x'], 'ro', color = 'darkgreen') #make it black #this is called format string #can do hax codes and add pound sign
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
#adobe illustrator, make legend, hue shifting or auto color you can change the colors from there

fig = plt.figure(figsize = (4, 5))

colors = { 'condition 1' : '#FF5733',
          'condition 2' : '#A44E3C',
          'condition 3' : '#599194',
          'condition 4' : '#3C9FA4',
          'condition 5' : '#1A4648'}
for i in range(5):
  plt.plot(np.cumsum(np.random.randn(1000)), color = colors[f'condition {i+1}'], label = f'condition {i+1}')
plt.legend()
plt.xlabel('xlabel', fontsize=14)
plt.xlim([0,500])
plt.title('title')

sns.despine(top=True, right=True) #remove the boarder of the box

fig.savefig('recording.pdf', bbox_inches = 'tight')
```

```python
fig , ax = plt.subplots(figsize = (15, 2.5), nrows=1, ncols=5, sharex=True, sharey=True) #five you figure and axis handle 
#sharex : set all sets equals the same axis

colors = { 'condition 1' : '#FF5733',
          'condition 2' : '#A44E3C',
          'condition 3' : '#599194',
          'condition 4' : '#3C9FA4',
          'condition 5' : '#1A4648'}
for i in range(5):
  plt.axes(ax[i]) #the current ax you're drawing
  plt.plot(np.cumsum(np.random.randn(1000)), color = colors[f'condition {i+1}'], label = f'condition {i+1}')
  plt.title(f'condition {i+1}')

plt.legend()
plt.xlabel('xlabel', fontsize=14)
plt.xlim([0,500])
plt.tight_layout() # making it tight the spacing between the plots


sns.despine(top=True, right=True) #remove the boarder of the box

fig.savefig('recording.pdf', bbox_inches = 'tight')
```

```python
#make use of seaborn example codes
#https://seaborn.pydata.org/examples/index.html
#bar plots are automaticaly put error bars 95% CI

import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
               palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```

```python
N = 1000
x = np.random.randn(N)
y = x+ 1.2 * np.random.randn(N)
plt.plot(x, y, 'ko')
```

```python
import pandas as pd
df = pd.DataFrame({'x': x, 'y': y})
df
```

```python
sns.lmplot(df, x='x', y='y', scatter_kws={'alpha': 0.3})
sns.despine(top=True, right=True)
#the error ribon is wrapping the regression line
```

```python
sns.regplot(df, x='x', y='y', scatter_kws={'alpha': 0.3})
sns.despine(top=True, right=True)
#it doesn't have a way to get out the coefficient value from the regression it plotted
# we need scipy, separate library to get the value of coefficient

```

```python
from scipy.stats import linregress
result = linregress(x,y)
print(result)
result.slope
```

```python
sns.regplot(df, x='x', y='y', scatter_kws={'alpha': 0.3})
sns.despine(top=True, right=True)
#it doesn't have a way to get out the coefficient value from the regression it plotted
# we need scipy, separate library to get the value of coefficient
slope = result.slope
intercept = result.intercept

reg_str = f'slope = {result.slope:0.3f}\nintercept = {result.intercept:0.3f}'
plt.text(0, 4, reg_str)
```
